## NMAP

Scanner activo pra indentificar los servicios host, servicios remotos, vulnerabilidades, puertos utilizados

Auditoria de puertos, pruebas de penetracion

Administradores de red utilizan NMAP para verificar la seguridad de sus redes

Envia sondas a muchos puertos TCP y analiza los que devuelven

REvisar el escript del NMAP

nmap - v scannme.nmap.org --------- Puertos abiertos con que servicios
nmap - v scannme.nmap.org -V -A  --- Sistema operativo

nmap -O 192.168.1.123 ---------------- Scanner un sitema operativo de una maquina

nmap -O 192.168.1.123 -T3 ------- (0-5) es para que vaya mas rapido y sea mas agresivo por si hay un sistema montorizando
